```python
import pandas as pd
import re
import string
#import spacy
from nltk.corpus import stopwords
```

```python
df=pd.read_csv('../data/cleaned_data.csv')
```

```python
df.head()
```

```python
data=df.copy()
```

```python
import nltk
nltk.download('stopwords')

# Charger le modèle de langue anglais de spaCy pour la lemmatisation
#nlp = spacy.load("en_core_web_sm")

# Créer un ensemble de mots vides en anglais
stop_words = set(stopwords.words('english'))
```

```python
# Fonction pour nettoyer le texte en supprimant les URLs, les mentions et la ponctuation
def clean_text(text):
    if isinstance(text, str):
        # Convertir en minuscule
        text = text.lower()

        # Supprimer les mentions
        text = re.sub(r'@\w+', '', text)

        # Supprimer la ponctuation
        text = text.translate(str.maketrans('', '', string.punctuation))

        # Supprimer les nombres
        text = re.sub(r'\d+', '', text)

        # Nettoyer les espaces supplémentaires
        text = text.strip()
        text = re.sub(r'\s+', ' ', text)
        
        return text
    else:
        return text

# Fonction de tokenisation et suppression des mots vides
def tokenize_and_remove_stopwords(text):
    if isinstance(text, str):
        # Tokeniser le texte
        tokens = text.split()
        
        # Supprimer les mots vides
        tokens = [word for word in tokens if word not in stop_words]
        
        return tokens
    else:
        return text
```

```python
data['Cleaned_Review'] = data['reviews.text'].apply(clean_text)

# Tokeniser et supprimer les mots vides
data['Tokenized_NoStopwords_review'] = data['Cleaned_Review'].apply(tokenize_and_remove_stopwords)
```

```python
data['Cleaned_title'] = data['reviews.title'].apply(clean_text)

# Tokeniser et supprimer les mots vides
data['Tokenized_NoStopwords_title'] = data['Cleaned_title'].apply(tokenize_and_remove_stopwords)
```
